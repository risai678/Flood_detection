# Flood Detection using Deep Learning 

```python
# Loading required libraries and functions
import os
import shutil
import random
import itertools
%matplotlib inline
import numpy as np
import tensorflow as tf
import matplotlib as mpl
from keras import backend
from tensorflow import keras
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
from IPython.display import Image
```

## Using finetuned model

```python
model = keras.models.load_model('model')
```

```python
# Prepare image for prediction

def preprocess_image(file):
    img_path = 'imagess/'
    img = image.load_img(img_path + file, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array_expanded_dims = np.expand_dims(img_array, axis=0)
    return tf.keras.applications.mobilenet.preprocess_input(img_array_expanded_dims)
```

```python
def get_prediction(imagee):
    preprocessed_image = preprocess_image(imagee)
    predictions = model.predict(preprocessed_image)
    result = np.argmax(predictions)
    # Print the predicted class label
    labels = ['Flooding', 'No Flooding']
    
    return labels[result]
    
```

```python
# Display image which we want to predict
Image(filename='imagess/5.jpeg', width=300,height=200) 
```

```python
# Preprocess image and make prediction
preprocessed_image = preprocess_image('5.jpeg')
predictions = model.predict(preprocessed_image)
```

```python
# Print predicted accuracy scores for both classes, i.e. (1) Flooding, (2) No Flooding
predictions
```

```python
# Get the maximum probability score for predicted class from predictions array
result = np.argmax(predictions)
```

```python
# Print the predicted class label
labels = ['Flooding', 'No Flooding']
labels[result]
```
